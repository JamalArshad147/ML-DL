```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
# Load Data

data = pd.read_csv('datasets/hotel_reviews.csv')
data.head()
```

```python
data['Review'][0]
```

```python
# Lowercase

data['review_lower'] = data['Review'].str.lower()
data['review_lower']
```

```python
# stopword removal 

en_stopwords = stopwords.words('english')
en_stopwords.remove('not')
print(en_stopwords)
```

```python
data['review_no_stopwords'] = data['review_lower'].apply(lambda x: " ".join([word for word in x.split() if word not in (en_stopwords)]))
```

```python
data.head()
```

```python
data['review_no_stopwords'][0]
```

```python
# punctuation

data['review_no_stopwords_no_punctuations'] = data.apply(lambda x: re.sub(r"[*]", 'star', x['review_no_stopwords']), axis=1)
```

```python
data.head()
```

```python
data['review_no_stopwords_no_punctuations'] = data.apply(lambda x: re.sub(r"[^\w\s]", '', x['review_no_stopwords_no_punctuations']), axis=1)
```

```python
data.head()
```

```python
# Tokenization

data['tokenize'] = data.apply(lambda x: word_tokenize(x['review_no_stopwords_no_punctuations']), axis=1)
```

```python
data.head()
```

```python
data['tokenize'][0]
```

```python
# Stemming

ps = PorterStemmer()
data["stemmed"] = data['tokenize'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
data.head()
```

```python
# Lemmatization

lemmatizer = WordNetLemmatizer()
data['lemmatised'] = data['tokenize'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
data.head()
```

```python
# N-Grams

token_clean =  sum(data['lemmatised'], [])
print(token_clean)
```

```python
unigram = (pd.Series(nltk.ngrams(token_clean, 1)).value_counts())
print(unigram)
```

```python
bigram = (pd.Series(nltk.ngrams(token_clean, 2)).value_counts())
print(bigram)
```

```python
trigram = (pd.Series(nltk.ngrams(token_clean, 3)).value_counts())
print(trigram)
```
